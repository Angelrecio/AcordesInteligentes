# Implementamos Knn

```python
import os
import librosa
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt

# Función para extraer características (espectrogramas) de archivos de audio
def extract_features(file_path, max_length=1000):
    y, sr = librosa.load(file_path)
    spectrogram = librosa.feature.melspectrogram(y=y, sr=sr)
    # Asegurar que todas las características tengan la misma longitud
    if spectrogram.shape[1] < max_length:
        pad_width = max_length - spectrogram.shape[1]
        spectrogram = np.pad(spectrogram, pad_width=((0, 0), (0, pad_width)), mode='constant', constant_values=0)
    else:
        spectrogram = spectrogram[:, :max_length]
    return librosa.power_to_db(spectrogram, ref=np.max).flatten()

# Ruta de la carpeta que contiene los archivos de entrenamiento
train_folder = "Training"
test_folder = "Test"

# Listas para almacenar características y etiquetas de entrenamiento
X_train = []
y_chord_train = []  # Etiqueta de acorde
y_play_style_train = []  # Etiqueta de estilo de tocar

# Recorrer la carpeta de entrenamiento para extraer características y etiquetas
for root, dirs, files in os.walk(train_folder):
    for file in files:
        if file.endswith(".wav"):
            file_path = os.path.join(root, file)
            # Obtener el nombre de la carpeta que contiene el archivo (nombre del acorde)
            chord_name = os.path.basename(root)
            play_style = file.split("_")[1].split(".")[0]
            y_chord_train.append(chord_name)
            y_play_style_train.append(play_style)
            # Extraer características (espectrograma) y añadirlo a la lista de características
            X_train.append(extract_features(file_path))

# Convertir listas a arrays numpy
X_train = np.array(X_train)
y_chord_train = np.array(y_chord_train)
y_play_style_train = np.array(y_play_style_train)

# Escalar características de entrenamiento
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)

# Dividir los datos en conjuntos de entrenamiento y prueba
X_train_chord, X_test_chord, y_chord_train, y_chord_test = train_test_split(
    X_train_scaled, y_chord_train, test_size=0.2, random_state=42
)
X_train_style, X_test_style, y_style_train, y_style_test = train_test_split(
    X_train_scaled, y_play_style_train, test_size=0.2, random_state=42
)

# Entrenar el clasificador KNN para predecir el acorde
knn_chord = KNeighborsClassifier(n_neighbors=3)
knn_chord.fit(X_train_chord, y_chord_train)

# Entrenar el clasificador KNN para predecir el estilo de tocar
knn_style = KNeighborsClassifier(n_neighbors=10)
knn_style.fit(X_train_style, y_style_train)

# Listas para almacenar características y etiquetas de prueba
X_test = []
y_chord_test = []  # Etiqueta de acorde
y_play_style_test = []  # Etiqueta de estilo de tocar

# Recorrer la carpeta de prueba para extraer características y etiquetas
for root, dirs, files in os.walk(test_folder):
    for file in files:
        if file.endswith(".wav"):
            file_path = os.path.join(root, file)
            # Obtener el nombre de la carpeta que contiene el archivo (nombre del acorde)
            chord_name = os.path.basename(root)
            play_style = file.split("_")[1].split(".")[0]
            y_chord_test.append(chord_name)
            y_play_style_test.append(play_style)
            # Extraer características (espectrograma) y añadirlo a la lista de características
            X_test.append(extract_features(file_path))

# Convertir listas a arrays numpy
X_test = np.array(X_test)
y_chord_test = np.array(y_chord_test)
y_play_style_test = np.array(y_play_style_test)

# Escalar características de prueba
X_test_scaled = scaler.transform(X_test)

# Predecir acordes para los datos de prueba usando KNN
y_chord_pred_test = knn_chord.predict(X_test_scaled)

# Predecir estilos de tocar para los datos de prueba usando KNN
y_play_style_pred_test = knn_style.predict(X_test_scaled)

# Calcular la precisión de la predicción en los datos de prueba para acordes
accuracy_chord_test = accuracy_score(y_chord_test, y_chord_pred_test)

# Calcular la precisión de la predicción en los datos de prueba para estilo de tocar
accuracy_play_style_test = accuracy_score(y_play_style_test, y_play_style_pred_test)

print("Precisión de predicción de acordes en datos de prueba:", accuracy_chord_test)
print("Precisión de predicción de estilo de tocar en datos de prueba:", accuracy_play_style_test)
```

```python
import os
import librosa
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

# Función para extraer características (espectrogramas) de archivos de audio
def extract_features(file_path, max_length=1000):
    y, sr = librosa.load(file_path)
    spectrogram = librosa.feature.melspectrogram(y=y, sr=sr)
    # Asegurar que todas las características tengan la misma longitud
    if spectrogram.shape[1] < max_length:
        pad_width = max_length - spectrogram.shape[1]
        spectrogram = np.pad(spectrogram, pad_width=((0, 0), (0, pad_width)), mode='constant', constant_values=0)
    else:
        spectrogram = spectrogram[:, :max_length]
    return librosa.power_to_db(spectrogram, ref=np.max).flatten()

# Ruta de la carpeta que contiene los archivos de entrenamiento
train_folder = "Training"
test_folder = "Test"

# Listas para almacenar características y etiquetas de entrenamiento
X_train = []
y_chord_train = []  # Etiqueta de acorde
y_play_style_train = []  # Etiqueta de estilo de tocar

# Recorrer la carpeta de entrenamiento para extraer características y etiquetas
for root, dirs, files in os.walk(train_folder):
    for file in files:
        if file.endswith(".wav"):
            file_path = os.path.join(root, file)
            # Obtener el nombre de la carpeta que contiene el archivo (nombre del acorde)
            chord_name = os.path.basename(root)
            play_style = file.split("_")[1].split(".")[0]
            y_chord_train.append(chord_name)
            y_play_style_train.append(play_style)
            # Extraer características (espectrograma) y añadirlo a la lista de características
            X_train.append(extract_features(file_path))

# Convertir listas a arrays numpy
X_train = np.array(X_train)
y_chord_train = np.array(y_chord_train)
y_play_style_train = np.array(y_play_style_train)

# Escalar características de entrenamiento
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)

# Dividir los datos en conjuntos de entrenamiento y prueba
X_train_chord, X_test_chord, y_chord_train, y_chord_test = train_test_split(
    X_train_scaled, y_chord_train, test_size=0.2, random_state=42
)
X_train_style, X_test_style, y_style_train, y_style_test = train_test_split(
    X_train_scaled, y_play_style_train, test_size=0.2, random_state=42
)

# Entrenar el clasificador SVM para predecir el acorde
svm_chord = SVC(kernel='rbf', C=10, gamma="scale")
svm_chord.fit(X_train_chord, y_chord_train)

# Entrenar el clasificador SVM para predecir el estilo de tocar
svm_style = SVC(kernel='rbf', C=10, gamma="scale")
svm_style.fit(X_train_style, y_style_train)

# Listas para almacenar características y etiquetas de prueba
X_test = []
y_chord_test = []  # Etiqueta de acorde
y_play_style_test = []  # Etiqueta de estilo de tocar

# Recorrer la carpeta de prueba para extraer características y etiquetas
for root, dirs, files in os.walk(test_folder):
    for file in files:
        if file.endswith(".wav"):
            file_path = os.path.join(root, file)
            # Obtener el nombre de la carpeta que contiene el archivo (nombre del acorde)
            chord_name = os.path.basename(root)
            play_style = file.split("_")[1].split(".")[0]
            y_chord_test.append(chord_name)
            y_play_style_test.append(play_style)
            # Extraer características (espectrograma) y añadirlo a la lista de características
            X_test.append(extract_features(file_path))

# Convertir listas a arrays numpy
X_test = np.array(X_test)
y_chord_test = np.array(y_chord_test)
y_play_style_test = np.array(y_play_style_test)

# Escalar características de prueba
X_test_scaled = scaler.transform(X_test)

# Predecir acordes para los datos de prueba usando SVM
y_chord_pred_test = svm_chord.predict(X_test_scaled)

# Predecir estilos de tocar para los datos de prueba usando SVM
y_play_style_pred_test = svm_style.predict(X_test_scaled)

# Calcular la precisión de la predicción en los datos de prueba para acordes
accuracy_chord_test = accuracy_score(y_chord_test, y_chord_pred_test)

# Calcular la precisión de la predicción en los datos de prueba para estilo de tocar
accuracy_play_style_test = accuracy_score(y_play_style_test, y_play_style_pred_test)

print("Precisión de predicción de acordes en datos de prueba:", accuracy_chord_test)
print("Precisión de predicción de estilo de tocar en datos de prueba:", accuracy_play_style_test)
```

```python
def predict_chords_with_sliding_window(audio_signal, svm_model, scaler, max_length=1000, window_overlap=0.5):
    """
    Predice los acordes utilizando una ventana deslizante sobre la señal de audio.

    Parameters:
        audio_signal (array): Señal de audio.
        svm_model (SVC): Modelo SVM entrenado para la clasificación de acordes.
        scaler (StandardScaler): Objeto StandardScaler ajustado utilizado para escalar las características.
        max_length (int): Longitud máxima de los datos de entrenamiento.
        window_overlap (float): Proporción de solapamiento entre segmentos adyacentes (0.0 - 1.0).

    Returns:
        array: Predicciones de acordes para cada segmento de la señal de audio.
    """
    # Definir el tamaño de la ventana deslizante
    window_size = max_length  # Longitud máxima de los datos de entrenamiento
    overlap = int(window_size * window_overlap)  # Solapamiento entre segmentos adyacentes

    # Dividir la señal de audio en segmentos con una ventana deslizante
    segments = []
    start = 0
    end = window_size
    while end <= len(audio_signal):
        segment = audio_signal[start:end]
        segments.append(segment)
        start += overlap
        end += overlap

    # Predicciones de acordes para cada segmento
    chord_predictions = []
    for segment in segments:
        # Extraer características del segmento de audio
        features = extract_features_window(segment, max_length)
        # Escalar características
        scaled_features = scaler.transform(features.reshape(1, -1))
        # Predecir el acorde usando el modelo SVM
        chord_prediction = svm_model.predict(scaled_features)
        chord_predictions.append(chord_prediction)

    return chord_predictions
```

```python
import os
import librosa
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
from keras.optimizers import RMSprop

# Función para extraer características (espectrogramas) de archivos de audio
def extract_features(file_path, max_length=1000):
    y, sr = librosa.load(file_path)
    spectrogram = librosa.feature.melspectrogram(y=y, sr=sr)
    # Asegurar que todas las características tengan la misma longitud
    if spectrogram.shape[1] < max_length:
        pad_width = max_length - spectrogram.shape[1]
        spectrogram = np.pad(spectrogram, pad_width=((0, 0), (0, pad_width)), mode='constant', constant_values=0)
    else:
        spectrogram = spectrogram[:, :max_length]
    return librosa.power_to_db(spectrogram, ref=np.max)

# Ruta de la carpeta que contiene los archivos de entrenamiento
train_folder = "Training"
test_folder = "Test"

# Listas para almacenar características y etiquetas de entrenamiento
X_train = []
y_train = []  # Etiqueta de acorde

# Recorrer la carpeta de entrenamiento para extraer características y etiquetas
for root, dirs, files in os.walk(train_folder):
    for file in files:
        if file.endswith(".wav"):
            file_path = os.path.join(root, file)
            # Obtener el nombre de la carpeta que contiene el archivo (nombre del acorde)
            chord_name = os.path.basename(root)
            y_train.append(chord_name)
            # Extraer características (espectrograma) y añadirlo a la lista de características
            X_train.append(extract_features(file_path))

# Convertir listas a arrays numpy
X_train = np.array(X_train)
y_train = np.array(y_train)

# Codificar etiquetas
label_encoder = LabelEncoder()
y_train_encoded = label_encoder.fit_transform(y_train)
num_classes = len(label_encoder.classes_)

# Dividir los datos en conjuntos de entrenamiento y prueba
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train_encoded, test_size=0.2, random_state=42)

# Expande las dimensiones de X_train y X_val para que coincida con el formato de entrada de CNN
X_train = np.expand_dims(X_train, axis=-1)
X_val = np.expand_dims(X_val, axis=-1)

# Normalizar los datos
X_train /= np.max(X_train)
X_val /= np.max(X_val)

# Definir la arquitectura de la red neuronal
model = Sequential()
model.add(Conv2D(32, (3, 3), activation='relu', input_shape=(X_train.shape[1], X_train.shape[2], 1)))
model.add(MaxPooling2D((2, 2)))
model.add(Conv2D(64, (3, 3), activation='relu'))
model.add(MaxPooling2D((2, 2)))
model.add(Conv2D(128, (3, 3), activation='relu'))
model.add(MaxPooling2D((2, 2)))
model.add(Flatten())
model.add(Dense(128, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(num_classes, activation='softmax'))

# Compilar el modelo con RMSprop optimizer
optimizer = RMSprop(learning_rate=0.0001)  # Adjust learning rate
model.compile(optimizer=optimizer, loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Entrenar el modelo con más epochs y un tamaño de lote más grande
model.fit(X_train, y_train, epochs=50, batch_size=64, validation_data=(X_val, y_val))

# Evaluación del modelo en los datos de prueba
X_test = []
y_test = []  # Etiqueta de acorde

# Recorrer la carpeta de prueba para extraer características y etiquetas
for root, dirs, files in os.walk(test_folder):
    for file in files:
        if file.endswith(".wav"):
            file_path = os.path.join(root, file)
            # Obtener el nombre de la carpeta que contiene el archivo (nombre del acorde)
            chord_name = os.path.basename(root)
            y_test.append(chord_name)
            # Extraer características (espectrograma) y añadirlo a la lista de características
            X_test.append(extract_features(file_path))

# Convertir listas a arrays numpy
X_test = np.array(X_test)
y_test = np.array(y_test)

# Codificar etiquetas
y_test_encoded = label_encoder.transform(y_test)

# Expande las dimensiones de X_test para que coincida con el formato de entrada de CNN
X_test = np.expand_dims(X_test, axis=-1)

# Normalizar los datos
X_test /= np.max(X_test)

# Evaluación del modelo en los datos de prueba
loss, accuracy = model.evaluate(X_test, y_test_encoded)
print("Accuracy on test data:", accuracy)
```
